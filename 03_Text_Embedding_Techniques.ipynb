```python
from flair.models import TextClassifier
from flair.data import Sentence

classifier = TextClassifier.load('en-sentiment')
```

```python
import pandas as pd
```

```python
sentences = [
    "I absolutely loved the new Batman movie",
    "The food was cold and tasteless",
    "It was an average experience, nothing special",
    "I am very excited to start my new job tomorrow",
    "The Phone is terrible I regret buying it",
    "The staff were friendly and helpful",
    "Its raining heavily, What a dull day",
    "The app keeps crashing Its so annoying",
    "Nothing went according to my plan today",
    "The equipment is operating, but occasionally it makes noise."
]
```

```python
res = []

for text in sentences:
    sent = Sentence(text)   # ✅ use text here
    classifier.predict(sent)
    
    label = sent.labels[0]
    
    res.append({            # ✅ correct method
        'Sentence': text,
        'Sentiment': label.value,
        'Confidence': round(label.score, 4)
    })

df = pd.DataFrame(res)
df
```

```python
# Question- Answering : Extracts exact answers from given context based on question.

from transformers import pipeline

qa = pipeline("question-answering")

context = "Apple Inc. has a net worth of over 2 trillion dollars."

question = "What is the net worth of Apple Inc?"

result = qa(
    context=context,
    question=question
)

print(result)
```

```python
# Zero - shot classifier: Classifies text into unseen labels without task-specific training.

classifier = pipeline('zero-shot-classification')

text = 'I want to return this item and get a refund'

labels = ['complaint', 'inquiry','feedback','spam']

result1 = classifier(text, labels)

print(result1)
```

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-mr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
# Use MarianMTModel instead of MarianModel for generation
model = MarianMTModel.from_pretrained(model_name)

text1 = "go and eat your food"

tokens = tokenizer(text1, return_tensors="pt", padding=True)
translated = model.generate(**tokens)

marathi_translate = tokenizer.decode(translated[0], skip_special_tokens=True)

print(marathi_translate)
```

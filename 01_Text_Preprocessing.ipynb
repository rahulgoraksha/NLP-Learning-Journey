## Text Preprocessing Techniques
Preprocessing is an important to clean and prepare the raw text data for analysis. Common preprocessing steps include:

* Tokenization
* Stopword Removal
* Punctuation Removal

* Lemmatization
* Text Normalization

* Parsing

Tokenization : 
used to in NLP to split paragraph and sentence into smaller unit that can be more easily assigned meaning
* Word Tokenization: Splits text into individual words or meaningful word units.
* Sentence Tokenization: Divides text into separate sentences based on punctuation marks.

Stopword Removal : Word that don't add meaning an are generaly filtered out before processing a natural language

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from string import punctuation
```

```python
list(punctuation)
```

```python
var = "Natural language processing (NLP) is the processing of natural language information by a computer. NLP is a subfield of computer science and is closely associated with artificial intelligence. NLP is also related to information retrieval, knowledge representation, computational linguistics, and linguistics more broadly."
```

```python
token = word_tokenize(var)

stop = set(stopwords.words('english'))
punct = set(punctuation)

filterword = [word for word in token if word.lower() not in stop and word not in punct]
print(filterword)
```

```python
for word in filterword:
    print(word)
    print(type(word))
```

```python
for word in filterword:
    print(word)
```

Word Sense Disambiguation: is an important method of NLP by which the meaning of a word is determine,  which is been used in a particular context

for eg: 
1. mouse is running
2. mouse is working

```python
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk
```

```python
x1 = "Sunrise (or sunup) is the moment when the upper rim of the Sun appears on the horizon in the morning,[1] at the start of the Sun path. The term can also refer to the entire process of the solar disk crossing the horizon."
```

```python
x2 = "A computer mouse (plural mice; also mouses)[nb 1] is a hand-held pointing device that detects two-dimensional motion relative to a surface. This motion is typically translated into the motion of the pointer (called a cursor) on a display, which allows a smooth control of the graphical user interface of a computer."
```

```python
l = lesk(word_tokenize(x1),"sun")
l.definition()
```

```python
l = lesk(word_tokenize(x2),"mouse")
l.definition()
```
